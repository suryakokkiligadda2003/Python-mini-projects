```python

#to get or fech data we can use requests
#get, post,set...
#it is also a function which enables the user performance with any server and any database source errors
import requests
```

```python
url = 'https://api.github.com/search/repositories'
```

```python
parms = {
         'q' : 'pandas+languages:python',
          'sort' : 'stars',
          'order' : 'desc'
}
```

```python
response = requests.get(url=url, params=parms)
```
