```python
#Function to get a graph
def getgraph(DataFrames,Column1,Column2):
    from matplotlib import pyplot as plt

    #Check the length of Column1 and Column2
    if len(DataFrames[Column1])==len(DataFrames[Column2]):
        DataFrames.plot(x= Column1,y=Column2,kind ='bar', color=['Black','skyblue'])
        print ('The graph/plot is successfully created ')
    else:
        print('please check the column size once')
```
